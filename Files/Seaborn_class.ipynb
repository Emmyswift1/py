```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('mpg.csv')
sns.scatterplot(x='cylinders',y='weight', data=df, hue='origin', hue_order=['europe','japan','usa'])
plt.show()
```

```python
hue_colors={'europe':'black','japan':'red','usa':'yellow'}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('mpg.csv')
sns.scatterplot(x='cylinders',y='weight', data=df, hue='origin', palette=hue_colors)
plt.show()
```
